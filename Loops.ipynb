```python
number=1
rows=5
for i in range(1, rows+1):
  for j in range(i):
   print(number,end=" ")
  print() # Moved here to create a new line after each row
  number=number+1
```

```python
ch='A'
rows=5
for i in range(1,rows+1):
  for j in range(i):
    print(ch,end=" ")
    ch=chr(ord(ch)+1)
  print()
```

```python
row=5
num=1
for i in range (1, row + 1):
  for j in range(i):
    print(num, end=" ")
    num+=1
  print() # Moved here to create a new line after each row
```

```python
n=6
# Top half of the diamond (increasing stars)
for i in range(1,n+1):
  print(" "*(n-i) + "*"*(2*i-1))

# Bottom half of the diamond (decreasing stars)
for i in range(n-1,0,-1):
  print(" "*(n-i) + "*"*(2*i-1))
```
